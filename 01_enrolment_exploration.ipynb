```python
import pandas as pd
import numpy as np

# Ensure date is datetime
df_enrol_clean["date"] = pd.to_datetime(df_enrol_clean["date"], dayfirst=True)

# Total enrolment per row
df_enrol_clean["total_enrolment"] = (
    df_enrol_clean["age_0_5"]
    + df_enrol_clean["age_5_17"]
    + df_enrol_clean["age_18_greater"]
)

# Age shares (proportions)
df_enrol_clean["share_0_5"] = df_enrol_clean["age_0_5"] / df_enrol_clean["total_enrolment"]
df_enrol_clean["share_5_17"] = df_enrol_clean["age_5_17"] / df_enrol_clean["total_enrolment"]
df_enrol_clean["share_18_plus"] = df_enrol_clean["age_18_greater"] / df_enrol_clean["total_enrolment"]
```

```python
state_date = (
    df_enrol_clean
    .groupby(["state", "region", "date"], as_index=False)
    .agg(
        daily_enrolment=("total_enrolment", "sum"),
        share_0_5=("share_0_5", "mean"),
        share_5_17=("share_5_17", "mean"),
        share_18_plus=("share_18_plus", "mean")
    )
)

state_date.head()
```

```python
state_metrics = (
    state_date
    .groupby(["state", "region"], as_index=False)
    .agg(
        avg_daily_enrolment=("daily_enrolment", "mean"),
        volatility=("daily_enrolment", "std"),
        coverage_days=("date", "nunique"),
        share_0_5=("share_0_5", "mean"),
        share_5_17=("share_5_17", "mean"),
        share_18_plus=("share_18_plus", "mean")
    )
)

state_metrics.head()
```

```python
ne_states_metrics = state_metrics[state_metrics["region"] == "NE"]

ne_reference = ne_states_metrics.describe()
ne_reference
```

```python
metrics_cols = [
    "avg_daily_enrolment",
    "volatility",
    "share_0_5",
    "share_5_17",
    "share_18_plus"
]

ne_centroid = ne_states_metrics[metrics_cols].mean()

mainland_states = state_metrics[state_metrics["region"] == "Mainland"].copy()

mainland_states["ne_similarity_score"] = mainland_states[metrics_cols].apply(
    lambda row: np.linalg.norm(row.values - ne_centroid.values),
    axis=1
)

mainland_states.sort_values("ne_similarity_score").head(10)
```

```python
mainland_median = mainland_states["avg_daily_enrolment"].median()

ne_outperformers = ne_states_metrics[
    ne_states_metrics["avg_daily_enrolment"] > mainland_median
]

ne_outperformers
```

```python
ne_states_metrics.sort_values("volatility").head()
```

```python
ne_volatility_upper = ne_states_metrics["volatility"].quantile(0.75)

volatile_mainland = mainland_states[
    mainland_states["volatility"] > ne_volatility_upper
].sort_values("volatility", ascending=False)

volatile_mainland.head()
```

```python
district_date = (
    df_enrol_clean
    .groupby(["state", "district", "date"], as_index=False)
    .agg(
        daily_enrolment=("total_enrolment", "sum")
    )
)

district_date.head()
```

```python
district_metrics = (
    district_date
    .groupby(["state", "district"], as_index=False)
    .agg(
        avg_daily_enrolment=("daily_enrolment", "mean"),
        volatility=("daily_enrolment", "std"),
        coverage_days=("date", "nunique")
    )
)

district_metrics.head()
```

```python
district_metrics["district_rank"] = (
    district_metrics
    .groupby("state")["avg_daily_enrolment"]
    .rank(method="dense", ascending=False)
)
```

```python
high_districts = district_metrics[
    district_metrics["district_rank"] <=
    district_metrics.groupby("state")["district_rank"].transform("max") * 0.1
]

high_districts.head()
```

```python
low_districts = district_metrics[
    district_metrics["district_rank"] >=
    district_metrics.groupby("state")["district_rank"].transform("max") * 0.9
]

low_districts.head()
```

```python
district_spread = (
    district_metrics
    .groupby("state")
    .agg(
        max_avg=("avg_daily_enrolment", "max"),
        min_avg=("avg_daily_enrolment", "min")
    )
    .assign(spread=lambda x: x["max_avg"] - x["min_avg"])
    .sort_values("spread", ascending=False)
)

district_spread.head(10)
```
